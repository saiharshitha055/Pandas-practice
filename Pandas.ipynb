Merging

```python
import pandas as pd
students=pd.DataFrame({
    "Name":["Meena","Nyra","Olivia","Gastro"],
    "Hometown":["Ongole","USA","UK","Australia"],
    "Age":[23,22,21,20],
    "Gender":["F","M","F","F"]
    })
books=pd.DataFrame({
    "Name":["Ramayana","Mahabharatam","Horror","Science"],
    "Hometown":["Ongole","USA","UK","Meerut"],
    "Age":[23,22,21,20],
    "Rating":["Good","Excellent","Fair","Very Good"]
})
students
```

```python
books
```

```python
pd.merge(students,books, on="Age")
```

```python
pd.merge(students,books,on="Age",how="inner")
```

```python
pd.merge(students,books,on="Age",how="left")
```

```python
pd.merge(students,books,on="Age",how="right")
```

```python
pd.merge(students,books,on="Age",how="outer")
```

```python
customers = pd.DataFrame({
    "Cust_Id":[101,102,103,104],
    "Name":["Ram", "Sam", "khushi", "Riya"]
})

orders = pd.DataFrame({
    "Cust_Id":[101,102,105],
    "Order":["Pizza", "Noodles", "Burger"]
})
```

```python
#inner merge
pd.merge(customers,orders,on="Cust_Id",how="inner")
```

```python
pd.merge(customers,orders,on="Cust_Id",how="left")
```

```python
pd.merge(customers,orders,on="Cust_Id",how="right")
```

```python
pd.merge(customers,orders,on="Cust_Id",how="outer")
```

One-hot Encoding

```python
df6 = pd.DataFrame({'cust_id':[101,102,103,104,105],
                   'age':[21,24,25,21,24],
                   "gender":["male",'female','male','female','female'],
                   "ratings":['excellent','good','average','average','good']})
```

```python
import pandas as pd
df7= pd.DataFrame({
    "Color": ["Red", "Blue", "Green", "Blue"]
})
encoded = pd.get_dummies(df7, columns=["Color"])
print(encoded)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
df8=pd.DataFrame({
    "Color":["Green","red","Blue","red"]
})
encoder = OneHotEncoder(sparse_output=False)
encoded = encoder.fit_transform(df8[["Color"]])
print(encoded)
```
